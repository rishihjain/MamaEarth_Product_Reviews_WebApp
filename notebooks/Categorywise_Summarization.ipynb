```python
import pandas as pd

price = pd.read_csv('Price.csv')
```

```python
price_grouped_df = price.groupby('PRODUCT_NAME')['REVIEW_CONTENT'].agg(lambda x: ' '.join(x)).reset_index()
price_grouped_df
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer
# Initialize T5 model and tokenizer
model = T5ForConditionalGeneration.from_pretrained("t5-base")
tokenizer = T5Tokenizer.from_pretrained("t5-base")
```

```python
# Function to generate summaries
def generate_summary(text):
    inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=512, truncation=True)
    summary_ids = model.generate(inputs, max_length=150, min_length= 3, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

# Group by product and join reviews
price_grouped_df = price.groupby('PRODUCT_NAME')['REVIEW_CONTENT'].agg(lambda x: ' '.join(x)).reset_index()

# Add a new column 'summary' with generated summaries
price_grouped_df['Price Summary'] = price_grouped_df['REVIEW_CONTENT'].apply(generate_summary)
```

```python
price_grouped_df
```

```python
data = pd.read_csv('Overall Service.csv')
```

```python
data = pd.merge(data, price_grouped_df[['PRODUCT_NAME', 'Price Summary']], on='PRODUCT_NAME', how='left')
```

```python
data.to_csv('Overall Service Price.csv', index = False)
```
