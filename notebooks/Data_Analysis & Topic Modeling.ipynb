## Sentiment Analysis of Reviews

```python
from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Perform sentiment analysis on each review
def get_sentiment(text):
    analysis = TextBlob(str(text))
    # Classify the polarity of the text as positive, negative, or neutral
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity < 0:
        return 'negative'
    else:
        return 'neutral'
```

```python
# Apply sentiment analysis to the 'REVIEW_CONTENT' column
df['sentiment'] = df['REVIEW_CONTENT'].apply(get_sentiment)

# Calculate the proportion of positive, negative, and neutral sentiments
sentiment_counts = df['sentiment'].value_counts()
total_reviews = len(df)

# Custom color palette
colors = ['#55a868', '#c44e52', '#4c72b0']

# Plot the sentiment distribution
plt.figure(figsize=(10, 8))
explode = (0.1, 0, 0)  # Explode the 'positive' slice for emphasis
sentiment_counts.plot(kind='pie', autopct='%1.1f%%', colors=colors, explode=explode, startangle=140)

# Draw a circle at the center to create a 'donut' chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Sentiment Distribution of Customer Reviews', fontsize=16)
plt.legend(sentiment_counts.index, loc='best', bbox_to_anchor=(0.8, 0.8))
plt.show()

# Display the proportion of each sentiment
print(f"Proportion of Positive Sentiments: {sentiment_counts['positive'] / total_reviews * 100:.1f}%")
print(f"Proportion of Negative Sentiments: {sentiment_counts['negative'] / total_reviews * 100:.1f}%")
print(f"Proportion of Neutral Sentiments: {sentiment_counts['neutral'] / total_reviews * 100:.1f}%")
```

```python
from gensim import corpora, models
from pprint import pprint
import nltk
from nltk import FreqDist
nltk.download('stopwords') # run this one time
import re
import spacy
# libraries for visualization
import pyLDAvis
#import pyLDAvis.gensim
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# function to plot most frequent terms
def freq_words(x, terms = 30):
  all_words = ' '.join([text for text in x])
  all_words = all_words.split()

  fdist = FreqDist(all_words)
  words_df = pd.DataFrame({'word':list(fdist.keys()), 'count':list(fdist.values())})

  # selecting top 20 most frequent words
  d = words_df.nlargest(columns="count", n = terms)
  plt.figure(figsize=(30,5))
  ax = sns.barplot(data=d, x= "word", y = "count")
  ax.set(ylabel = 'Count')
  plt.show()
```

```python
# remove unwanted characters, numbers and symbols
df['REVIEW_CONTENT'] = df['REVIEW_CONTENT'].str.replace("[^a-zA-Z#]", " ")
```

```python
#Let’s try to remove the stopwords and short words (<2 letters) from the reviews.

from nltk.corpus import stopwords
stop_words = stopwords.words('english')
# function to remove stopwords
def remove_stopwords(rev):
    rev_new = " ".join([i for i in rev if i not in stop_words])
    return rev_new

# remove short words (length < 3)
df['REVIEW_CONTENT'] = df['REVIEW_CONTENT'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>2]))

# remove stopwords from the text
reviews = [remove_stopwords(r.split()) for r in df['REVIEW_CONTENT']]

# make entire text lowercase
reviews = [r.lower() for r in reviews]
```

```python
freq_words(reviews, 35)
```

```python
# To further remove noise from the text we can use lemmatization from the spaCy library.
#It reduces any given word to its base form thereby reducing multiple forms of a word to a single word.

!python -m spacy download en # one time run
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])

def lemmatization(texts, tags=['NOUN', 'ADJ']): # filter noun and adjective
       output = []
       for sent in texts:
             doc = nlp(" ".join(sent))
             output.append([token.lemma_ for token in doc if token.pos_ in tags])
       return output
```

```python
# Let’s tokenize the reviews and then lemmatize the tokenized reviews.

tokenized_reviews = pd.Series(reviews).apply(lambda x: x.split())
print(tokenized_reviews[1])
```

```python
reviews_2 = lemmatization(tokenized_reviews)
print(reviews_2[1]) # print lemmatized review
```

```python
reviews_3 = []
for i in range(len(reviews_2)):
    reviews_3.append(' '.join(reviews_2[i]))

df['reviews'] = reviews_3

freq_words(df['reviews'], 35)
```

```python
dictionary = corpora.Dictionary(reviews_2)
```

```python
dictionary.filter_extremes(no_below=5, no_above=0.5, keep_n=2000)
print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(rev) for rev in reviews_2]
```

```python
# Define a function to build and evaluate LDA models
def evaluate_lda_model(corpus, dictionary, num_topics, passes):
    lda_model = models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=passes)
    coherence_model = models.CoherenceModel(model=lda_model, texts=tokenized_reviews, dictionary=dictionary, coherence='c_v')
    coherence_score = coherence_model.get_coherence()
    return lda_model, coherence_score

# Experiment with different forms of LDA
num_topics_list = [3, 5,7]
passes_list = [10, 15, 20]

best_lda_model = None
best_coherence_score = -1

for num_topics in num_topics_list:
    for passes in passes_list:
        lda_model, coherence_score = evaluate_lda_model(corpus, dictionary, num_topics, passes)
        print(f"Num Topics: {num_topics}, Passes: {passes}, Coherence Score: {coherence_score}")

        if coherence_score > best_coherence_score:
            best_lda_model = lda_model
            best_coherence_score = coherence_score

# Print the topics and associated words of the best model
pprint(best_lda_model.print_topics())
```
